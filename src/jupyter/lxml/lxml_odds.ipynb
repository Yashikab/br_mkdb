```python
import lxml.html as lxml
```

```python
with open("odds_3fuku_2020040869.html", "r") as f:
    html_source = f.read()
```

```python
lx_content = lxml.fromstring(html_source)
```

```python
table_xpath = "/html/body/main/div/div/div/div[2]/div[6]/table/tbody/tr/td[contains(@class, 'oddsPoint')]"
```

```python
pay_list = lx_content.xpath("/".join([table_xpath]))
```

```python
renfuku_key_list = []
for trd in range(6, 0, -1):
    for snd in range(trd-1, 0, -1):
        for fst in range(snd-1, 0, -1):
            renfuku_key_list.append(f'{fst}-{snd}-{trd}')
```

```python
renfuku_key_list
```

```python
odds = list(map(lambda x: float(x.text), pay_list))
```

```python
odds.reverse()
```

```python
odds
```

```python
/html/body/main/div/div/div/div[2]/div[6]/table/tbody/tr[2]/td[12]
```

```python
content_rv_dict = {}
for keyname in renfuku_key_list:
    content_rv_dict[keyname] = odds.pop(0)
```

```python
# キーを並び替え
content_dict = {}
for key, value in sorted(content_rv_dict.items(), key=lambda x:x[0]):
    content_dict[key] = value
```
